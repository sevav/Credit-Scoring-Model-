```python
from fastai.tabular import *
from sklearn import preprocessing

```

```python
dftrain = pd.read_csv('./data/training_features_proc.csv')
dftest = pd.read_csv('./data/testing_features_proc.csv')


dftrain = pd.DataFrame(dftrain.dropna())
dftest = pd.DataFrame(dftest.dropna())
dftrain.head()
```

```python
dftest.head()
```

```python
cols = dftrain.columns
```

```python
x = dftrain.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
x_scaled = min_max_scaler.fit_transform(x)
y_scaled = min_max_scaler.fit_transform(dftest)
dftrain_scaled = pd.DataFrame(x_scaled, columns=cols)
dftest_scaled = pd.DataFrame(y_scaled, columns=cols)
```

```python
dftrain_scaled.head()
dftest_scaled.head()
```

```python
cols = dftrain_scaled.columns
cols
```

```python
dep_var = 'Credit Score'
cont_names = cols.drop('Credit Score')
cont_names
```

```python
test = TabularList.from_df(dftest_scaled, cont_names=cont_names)
```

```python
data = (TabularList.from_df(dftrain_scaled, cont_names=cont_names)
                           .split_by_idx(list(range(800,1000)))
                           .label_from_df(cols=dep_var)
                           .databunch())
```

```python
data.show_batch(rows=10)
```

```python
learn = tabular_learner(data, layers=[200,50], metrics= root_mean_squared_error, wd=0.01)
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(4, 1e-2)
```
